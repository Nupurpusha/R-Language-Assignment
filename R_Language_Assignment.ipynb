```r
generate_random_string <- function(length = 10) {
  letters <- c(letters, LETTERS, 0:9)
  paste0(sample(letters, length, replace = TRUE), collapse = "")
}
random_strings <- replicate(1000, generate_random_string())
file_path <- "random_strings.txt"
writeLines(random_strings, file_path)
file_path
```
